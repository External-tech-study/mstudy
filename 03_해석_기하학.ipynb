### 유클리디안 놈

- 벡터의 길이 또는 크기를 측정하는 방법
- 직선 거리

<br/>

$n\text{차원 유클리디안 공간에서 벡터 } \mathbf{v} = (v_1, v_2, \ldots, v_n) \text{의
유클리디안 놈은 다음과 같이 표현:}$

$\| \mathbf{v} \|_2 = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$

<br/>


## dot product

- 벡터의 요소를 곱한 후, 합산 i.e. 결과가 scala
- 두 벡터가 이루는 각도를 찾기 위해 사용

<br/>

<b>정의</b>

$ \mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + \cdots + a_nb_n$


<br/>


## inner product

- 벡터 공간에서 두 벡터의 유사도를 측정하는데 사용
- dot product의 일반화 버전

<br/>

inner product는 다음의 성질을 만족해야한다.

$ 1. \textbf{양의 정부호성}: (\langle \mathbf{a}, \mathbf{a} \rangle \geq 0)이며, (\langle \mathbf{a}, \mathbf{a} \rangle = 0)은 \ (\mathbf{a})가 \ 영벡터일 \ 때만 \ 성립 $


$ 2. \textbf{선형성}: (\langle \alpha \mathbf{a} + \beta \mathbf{b}, \mathbf{c} \rangle = \alpha \langle \mathbf{a}, \mathbf{c} \rangle + \beta \langle \mathbf{b}, \mathbf{c} \rangle) $


$ 3. \textbf{켤레 대칭성}: (\langle \mathbf{a}, \mathbf{b} \rangle = \overline{\langle \mathbf{b}, \mathbf{a} \rangle}) $


<br/>

> dot product는 유클리디안 공간에서 정의된 특수한 형태의 inner product

<br/>


### symmetric, positive definite matrices

- 대칭성과 양의 정부호성을 만족
- 행렬의 고유값 문제, 선형 시스템의 해법, 목적함수의 곡률등 여러 곳에서 응용

<br/>

$\forall \mathbf{x} \in V \setminus \{0\} : \mathbf{x}^\top A \mathbf{x} > 0.$

- 모든 x에 대하여, 위의 식이 성립해야함

<br/>

실수 행렬, <b>A</b>가 대칭이고 양의 부호성을 띈다면 다음을 성질을 만족한다.
- <b>A</b>의 널공간은 오직 0으로 구성.
- <b>A</b>의 대각 요소는 모두 양의 부호

<br/>

## 코시-슈바르츠 부등식

- 두 백터의 내적과 그 백터의 놈 사이의 관계를 설명

$ |\langle \mathbf{u}, \mathbf{v} \rangle| \leq \|\mathbf{u}\| \|\mathbf{v}\|$

<br>

## distance

- 두 점 사이의 길이 또는 간격
- inner product 로써, dot product가 사용됐다면, 해당 거리는 유클리디안 거리라고 부름

<br/>

## metric

$ 집합 \ ( X )와 \ 함수 \ ( d : X \times X \to \mathbb{R} )가 \ 있을 \ 때, \ ( d )가 \ 메트릭이 \ 되기 \ 위한 \ 조건은 \ 다음과 \ 같음 $

$ 1. \textbf{비음성 (Non-negativity)}: 모든 ( x, y \in X )에 \ 대해 \ ( d(x, y) \geq 0 ) $
$ 2. \textbf{동일성 (Identity of Indiscernibles)}: 모든 \  ( x, y \in X )에 \ 대해 ( d(x, y) = 0 )인 \ 경우는 \ 오직 ( x = y ) $
$ 3. \textbf{대칭성 (Symmetry)}: 모든 \ ( x, y \in X )에 \ 대해 ( d(x, y) = d(y, x) ) $
$ 4. \textbf{삼각 부등식 (Triangle Inequality)}: 모든 \ ( x, y, z \in X )에 \ 대해 ( d(x, y) + d(y, z) \geq d(x, z) )$

$이 \ 네 \ 가지 \ 조건을 \ 만족하는 \ 함수 \ ( d )를 \ 메트릭이라고 \ 하며, \ ( (X, d) )를 \ 메트릭 \ 공간이라고 부름$


<br/>

## Angle and Orthogonality


### angle
- 코시-슈바르츠부터 각도를 구할 수 있음


$ \cos \theta = \frac{\langle \mathbf{u}, \mathbf{v} \rangle}{\|\mathbf{u}\| \|\mathbf{v}\|} $


### orthgonality

- inner product의 결과가 0이라면, 피연산자로 사용된 2개의 벡터는 서로 직교한다(orthgonal)
- 0 벡터는 모든 벡터에 직교

<br/>


### orthogonal matrix
정방 행렬과 그 행렬의 전치의 곱으로 나오는 결과가 단위 행렬일 때, 그 행렬을 직교 행렬이라고 부른다.
- <b>A</b>전치 행렬 == <b>A</b>의 역행렬

<br/>

## Orthgonal basis

- 직교를 만족하는 기저 벡터
  - 자기 자신과의 inner product의 결과는 1
  - 서로 다른 벡터에 대한 inner product 결과는 0

<br/>

## Orthogonal complement(직교 여공간)

$ 벡터 공간 \ ( V )의 \ 부분 \ 공간 \ ( W )에 \ 대한 \ 직교 \ 여공간 \ ( W^\perp )는 \ ( V )의 \ 모든 \ 벡터 \ 중에서 \ ( W )의 \ 모든 \ 벡터에 \ 직교하는 \ 벡터들로 \ 이루어진 \ 집합 $

$ W^\perp = \{ \mathbf{v} \in V \mid \langle \mathbf{v}, \mathbf{w} \rangle = 0 \text{ for all } \mathbf{w} \in W \} \ $

- $ 여기서 \ ( \langle \mathbf{v}, \mathbf{w} \rangle )는 \ 벡터 \ ( \mathbf{v} )와 \ ( \mathbf{w} )의 \ 내적 $


<br/>

- ML에서는 고차원의 데이터를 저차원으로 변환 할 때, 정보 손실을 최소화 해야함
